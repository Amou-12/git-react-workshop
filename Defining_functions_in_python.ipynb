```python
#Excerscise 1
#Convert Temperature Write a function that converts a temperature from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):

    #### Write your function here
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit


    #### your function ends here

 # Test the function
temp_fahrenheit = celsius_to_fahrenheit(25)
print(f"Temperature in Fahrenheit: {temp_fahrenheit}")


#Excercise 2
#Check Even or Odd Write a function that checks whether a given number is even or odd.
def is_even(number):
    #### Write your function here

    if number % 2 == 0:
        return True   # The number is even
    else:
        return False  # The number is odd
    #### your function ends here


# Test the function
num = int(input("Enter a number: "))

if is_even(num):
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")


#Excerccise 3
#Find Maximum Write a function that returns the maximum of two numbers.
def find_maximum(num1, num2):
    #### Write your function here
    if num1 > num2:
        return num1
    else:
        return num2
    #### your function ends here


# Ask the user for two numbers
a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))

# Call the function and display the result
print(f"The maximum of {a} and {b} is {find_maximum(a, b)}")


#Excercise 4
#Calculate Factorial Write a function that calculates the factorial of a given number.
def factorial(n):
    #### Write your function here
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
    #### your function ends here


# Ask the user for a number
num = int(input("Enter a number to find its factorial: "))

# Call the function and display the result
print(f"The factorial of {num} is {factorial(num)}")

#Excercise 5
#Reverse a String Write a function that reverses a given string.
def reverse_string(text):
    #### Write your function here
    reversed_text = text[::-1]
    return reversed_text
    #### your function ends here


# Ask the user to enter a string
user_input = input("Enter a string: ")

# Call the function and display the result
print("Reversed string:", reverse_string(user_input))

#Excercise 6
#Count Vowels in a String Write a function that counts the number of vowels in a given string.
def count_vowels(text):
  vowels="aeiouAEIOU"
  count=0
  for char in text:
    if char in vowels:
      count+=1
  return count

  # Ask the user to enter a string
user_input = input("Enter a string: ")

# Call the function and display the result
print(f"Number of vowels: {count_vowels(user_input)}")



```
